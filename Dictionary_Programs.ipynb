### 1.Create a dictionary

```python
dic={
    "name":"anjali",
    "age":21,
    "course":"BCA"
}
print(dic)
```

### 2. Traverse dictinary

```python
student={
    "name":"anjali",
    "age":21,
    "course":"BCA"
}
for key,value in student.items():
    print(key,":",value)
```

### 3. Find key with maximum value 

```python
marks={
    "Math":80,
    "English":75,
    "Science":90
}
max_key=max(marks,key=marks.get)
print(max_key)
```

### 4. Sort dictionary by value

```python
marks={
    "Math":80,
    "English":75,
    "Science":90
}
sorted_dict=dict(sorted(marks.items(),key=lambda x:x[1]))
print(sorted_dict)
```

### 5. Merge two dictionaries

```python
d1={"a":1,"b":2}
d2={"c":3,"d":4}
d1.update(d2)
print(d1)
```

### 6. Student marks dictionary system

```python
marks={
    "Math":80,
    "English":75,
    "Science":90
}
total=sum(marks.values())
per=total/len(marks)
print("total marks:",total)
print("percentage: ",per)
```

### 7. remove a key

```python
marks={
    "Math":80,
    "English":75,
    "Science":90
}
print(marks.pop("Math"))
print(marks)
```
