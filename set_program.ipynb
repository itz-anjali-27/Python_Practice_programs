### 1. Create a set display

```python
s={1,2,3,4}
print(s)
```

### 2. Add element to a set

```python
s={1,2,3}
s.add(4)
print(s)
```

### 3. Add multiple element

```python
s={1,2}
s.update([3,4,5])
print(s)
```

### 4. remove element from set

```python
s={1,2,3}
s.remove(2)
s.discard(3)
print(s)
```

### 5. Union of two sets

```python
a={1,2,3}
b={3,4,5}
print(a.union(b))
```

### 6. Intersection of sets

```python
a={1,2,3}
b={3,4,5}
print(a.intersection(b))
```

### 7. Symmetric difference

```python
a={1,2,3}
b={3,4,5}
print(a.symmetric_difference(b))
```

### 8. check subset, superset, disjoint sets

```python
a={1,2}
b={1,2,3}
print(a.issubset(b))
print(a.issuperset(b))
print(a.isdisjoint(b))
```

### 9. Count unique elements from list

```python
lst=[1,2,3,3,3,2,2]
s=set(lst)
print(len(s))
```
