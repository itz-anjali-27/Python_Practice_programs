### 1. Count Vowels in a string

```python
str="hello worlds"
count=0
vowels="aeiouAEIOU"
for i in str:
    if i in vowels:
        count+=1
print(count) 
```

### 2.Reverse a string

```python
str=input("enter string : ")
print("Reverse a string : ",str[::-1])
```

### 3.Check palindrome string

```python
str=input("enter string : ")
s=str[::-1]
if(s==str):
    print("Palindrome")
else:
    print(" Not Palindrome")
```

### 4.Count words in a string

```python
str=input("enter string : ")
word=str.split()
print("Word count: ",len(word))
```

### 5. Count frequency of each character

```python
str=input("enter string : ")
freq={}

for ch in str:
    freq[ch]=freq.get(ch,0)+1

print(freq)

```

### 6. Find longest word in a sequence

```python
str=input("enter string : ")
word=str.split()
longest=max(word,key=len)
print("Longest word: ",longest)
```

### 7. Remove duplicate characters

```python
str=input("enter string : ")
result=""
for ch in str:
    if ch not in result:
        result+=ch

print("After removing duplicates : ",result)
```

### 8. Replace space with underscore

```python
s=input("Enter string : ")
print(s.replace(" ","_"))
```

### 9. Check anagram

```python
s1=input("Enter first string : ")
s2=input("Enter second string : ")

if sorted(s1)==sorted(s2):
    print("Anagram")
else:
    print("Not Anagram")
```
