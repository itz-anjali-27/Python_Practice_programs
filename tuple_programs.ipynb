### 1. tuple create and print

```python
t=(10,20,30,40)
print(t)
```

### 2. tuple length

```python
t=(10,20,30,40)
print(len(t))
```

### 3. tuple slicing

```python
t=(10,20,30,40)
print(t[0:5])
print(t[0:])
print(t[:3])
```

### 4. Find maximum and minimum

```python
t=(10,20,30,40)
print("max : ",max(t))
print("min : ",min(t))
```

### 5. Count element frequency

```python
t=(10,20,20,30,40)
print(t.count(20))
```

### 6. Find index of element

```python
t=(10,20,20,30,40)
print(t.index(20))
```

### 7. tuple unpacking

```python
t=(10,20,20)
a,b,c=t
print(a,b,c)
```

### 8. convert list to tuple

```python
l=[1,2,4]
t=tuple(l)
print(t)
```

### 9. Sum of tuple

```python
t=(10,20,20)
print(sum(t))
```

### 10. Reverse of tuple

```python
t=(10,20,20)
print(t[::-1])
```
