### 1. Create class and object 

```python
class Person:
    name="Ramesh"
p=Person()
print(p.name)
```

### 2. Constructor

```python
class Person:
    def __init__(self):
        self.name="anju"
        self.age=20
p=Person() 
print(p.name,p.age)
```

### 3. Student class with marks

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def display(self):
        print("Name : ",self.name)
        print("Marks : ",self.marks)

s=Student("Anjali",85)
s.display()
```

### 4. Method inside class

```python
class Calculator:
    def add(self,a,b):
        return a+b

c=Calculator()
print("sum : ",c.add(10,20))
```

### 5. Inheritance example

```python
class Parent:
    def show(self):
        print("This is parent class")

class Child(Parent):
    def display(self):
        print("This is child class")

c=Child()
c.show()
c.display()
        
```

### 6. Encapsulation example

```python
class Bank:
    def __init__(self):
        self.__balance=1000
    def show_balance(self):
        print("Balance : ",self.__balance)

b=Bank()
b.show_balance()
```

### 7. Method overriding

```python
class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):
        print("Child method")

c=Child()
c.show()
    
```

### 8. Method Overloading

```python
class Parent:
    def add(self,a,b=0,c=0):
        return a+b+c
    def add(self,a,b=0,c=0):
         return a+b+c    
a=Parent()
print(a.add(2,4))
print(a.add(2,4,5))
```

### 9. Operator overloading

```python
class Complex:
    def __init__(self,real,img):
        self.real=real
        self.img=img

    def show(self):
        print(self.real ,"+",self.img," i")

    def __add__(self,num2):
         self.Real=self.real+num2.real
         self.Img=self.img+self.img
         return f"{self.Real} + {self.Img} i"

c1=Complex(4,2) 
c1.show()
c2=Complex(4,2) 
c2.show()
c3=c1+c2
print(c3)
```

### 10. Static method

```python
class Student:
    @staticmethod
    def is_pass(marks):
        if marks>=33:
            return "Pass"
        else:
            return "Fail"

print(Student.is_pass(75))
print(Student.is_pass(20))
```

### 11. Super Keyword

```python
class Person:
    def __init__(self,name):
        self.name=name

class Student(Person):
    def __init__(self,name,marks):
        super().__init__(name)
        self.marks=marks
    def display(self):
        print(self.name, self.marks)

s=Student("Anjali",90)
s.display()
```

### 12. Abstract class

```python
from abc import ABC,abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self,side):
        self.side=side

    def area(self):
        print("Area : ",self.side*self.side)

s=Square(4)
s.area()
        
    
```
