### 1. write data into file

```python
f=open("data.txt","w")
f.write("hello python")
f.close()
print("Data written successfully")
```

### 2. Read data from file

```python
f=open("data.txt","r")
data=f.read()
print(data)
f.close()
```

### 3. Count number of lines in file

```python
f=open("data.txt","r")
data=f.readlines()
print(len(data))
f.close()
```

### 4. Counts words in file

```python
f=open("data.txt","r")
data=f.read()
words=data.split()
print(len(words))
f.close()
```

### 5.Count characters in file

```python
f=open("data.xt","r")
data=f.read()
print(len(data))
f.close()
```

### 6. Append data into file

```python
f=open("data.xt","a")
f.write("python is used in DS")
f.close()
print("Data written successfully")
```
