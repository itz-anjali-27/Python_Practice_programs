### 1.Print number from 1 to n

```python
n=int(input("enter number : "))
for i in range(1,n+1):
    print(i)
```

### 2.Sum of digits of a number

```python
n=int(input("enter number : "))
s=0
while(n>0):
    r=n%10
    s=s+r
    n=n//10
print("sum of digits : ",s)    
    
```

### 3. Reverse a number

```python
n=int(input("enter number : "))
s=0
while(n>0):
    r=n%10
    s=s*10+r
    n=n//10
print("Reverse : ",s)    
```

### 4. Check palindrom number

```python
n=int(input("enter number : "))
s=0
cn=n
while(n>0):
    r=n%10
    s=s*10+r
    n=n//10
if(cn==s):
    print("It is palindrom")
else:
    print("It is not palindrom")
```

### 5. Count digit of number

```python
n=int(input("enter number : "))
count=0
while(n>0):
    count+=1
    n//=10
print("Total digits : ",count)    
```

### 6. Print  Fibonacci series

```python
n=int(input("enter terms : "))
a=0
b=1
print(a)
print(b)
for i in range(n):
    c=a+b
    print(c)
    a=b
    b=c
```

### 7.Factorial of a number

```python
n=int(input("enter number : "))
fact=1
for i in range(1,n+1):
    fact*=i
print("Factorial: ",fact)
```

### 8. Check a prime number

```python
n=int(input("enter number : "))
isPrime=True
for i in range(2,n):
    if(n%i==0):
        isPrime=False
        break
if(isPrime):
    print("Prime number")
else:
    print("not prime number")
    
    
```

### 10. Armstrong number check

```python
n=int(input("enter number : "))
s=0
cn=n
while(n>0):
    r=n%10
    s=s+r*r*r
    n=n//10
if(cn==s):
    print("It is Armstrong")
else:
    print("It is not Armstrong")
```
