### 1. Find sum of list element

```python
l=[1,2,3,4,5,6]
s=0
for i in l:
    s+=i
print("sum of list: ",s)
```

### 2. Find max and min in list

```python
l=[1,2,3,4,5,6]
print("maximum : ",max(l))
print("minimum : ",min(l))
```

### 3. Remove duplicates from list

```python
list=[1,1,2,3,2,3,4,5,6]
new_list=[]
for i in list:
    if i not in new_list:
        new_list.append(i)
print(new_list)        
        
```

### 4. Sort list without using sort()

```python
l=[32,23,56,43,22]
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if(l[i]>l[j]):
            l[i],l[j]=l[j],l[i]
           
print(l)
```

 ### 5.Find Second largest number

```python
l=[32,23,56,43,22]
largest=max(l)
second=-9999

for i in l:
    if i!=largest and i>second:
        second=i
print("Second largest number is: ",second)
```

### 6. Count even & odd numbers in list

```python
l=[32,23,56,43,22]
e=0
o=0
for i in l:
    if i%2==0:
        e+=1
    else:
        o+=1
print("even : ",e)
print("odd : ",o)
```

### 7. Reverse list

```python
l=[32,23,56,43,22]
rev=[]
for i in range(len(l)-1,-1,-1):
    rev.append(l[i])
print("reverse : ",rev)
    
```

### 8.Merge two list

```python
l1=[32,23,56,43,22]
l2=[32,23,56,43,22]
merge=l1+l2
print(merge)
```

### 9. Find common element of two list

```python
l1=[32,23,56,43,22]
l2=[32,23,56,43,22]
common=[]

for i in l1:
    if i in l2:
        common.append(i)
print(common)
```
