### 1. fuction to find mean

```python
def mean(a,b):
    return (a+b)/2
print("mean : ",mean(3,7))    
    
```

### 2. find max of 3 numbers

```python
def max_num(a,b,c):
    return max(a,b,c)
print("maximum : ",max_num(3,7,8))    
    
```

### 3. check prime number

```python
def check_prime(n):
    isPrime=True
    for i in range(2,n):
        if n%i==0:
            isPrime=False
    if(isPrime):
        print("Prime number")
    else:
        print("Not prime number")

n=int(input("enter number : "))
check_prime(n)
            
```

### 4.Recursive factorial

```python
def fact(n):
    if n==0 or n==1:
        return 1
    return n*fact(n-1)
n=int(input("enter number : "))
print(fact(n))
    
```

### 5.Recursive Fibonacci

```python
def fibo(n):
    if n<=1:
        return n
    return fibo(n-1)+fibo(n-2)
n=int(input("enter number : "))
print(fibo(n))
    
```

### 6. Fuction to count vowels

```python
def count_vowels(s):
    vowels="aeiouAEIOU"
    count=0
    for ch in s:
        if ch in vowels:
            count+=1
    return count

print(count_vowels("Python"))    
```

### 7. Lambda function

```python
square=lambda x:x*x
n=int(input("enter number : "))
print(square(n))
```

### 8. Function returning multiple values

```python
def calculate(a,b):
    return a+b,a-b,a*b

add,sub,mul=calculate(10,5)
print(add,sub,mul)
```

### 9. Fuction to calculate average of list

```python
def average(list):
    return sum(list)/len(list)

print(average([10,20,30,40]))
```
