## Common buit-in exception

### 1. ValueError

```python
try:
    x=int("abc")
except ValueError:
    print("Value error")
```

### 2. IndexError

```python
try:
    x=[1,2,3]
    print(x[6])
except IndexError:
    print("Index error")
```

### 3. TypeError

```python
try:
    print(10+"2")
except TypeError:
    print("Type error")
```

### 4. FileNotFoundError

```python
try:
    f=open("data.txt")
except FileNotFoundError:
    print("file not found error")
```

### 5. ZeroDivisionError

```python
try:
    a=9
    b=0
    print(a/b)
except ZeroDivisionError:
    print("ZeroDivisionError error")
```

### # try-except-else 

```python
try:
    a=9
    b=3
    print(a/b)
except ZeroDivisionError:
    print("ZeroDivisionError error")
else:
    print("Division successful")
```

### # try-except-finally

```python
try:
    a=9
    b=3
    print(a/b)
except ZeroDivisionError:
    print("ZeroDivisionError error")
finally:
    print("Program finished")
```

### Multiple except blocks

```python
try:
    a=9
    b=0
    print(a/b)
except ZeroDivisionError:
    print("ZeroDivisionError error")
except ValueError:
    print("invaild value")
```

```python
class AgeError(Exception):
    pass
try:
    age=int(input("Enter age : "))
    if age<18:
        raise AgeError("Not eligible")
    print("Eligible")    
except AgeError as e:
    print(e)
```
